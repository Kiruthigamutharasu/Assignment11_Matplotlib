```python
# Importing Matplotlib
import matplotlib.pyplot as plt
import csv
```

```python

# Step 1 : Load Kaggle Dataset using CSV

category = []
sales = []
profit = []
quantity = []
region = []
sub_category = []

with open("SampleSuperstore.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)

    for row in reader:
        category.append(row["Category"])
        sub_category.append(row["Sub-Category"])
        sales.append(float(row["Sales"]))
        profit.append(float(row["Profit"]))
        quantity.append(int(row["Quantity"]))
        region.append(row["Region"])
```

```python
#TASK 1 : LINE PLOT (Sales Trend)
# Task 1 - Step 1 : Create Line Plot sales trend over months
plt.plot(sales[:12])

# Task 1 - Step 2 : Add Title and Labels
plt.title("Sales Trend")
plt.xlabel("Index")
plt.ylabel("Sales")

# Task 1 - Step 3 : Display Plot
plt.show()
```

```python
# TASK 2 : SCATTER PLOT

# Task 2 - Step 1 : Create Scatter Plot
plt.scatter(sales[:50], profit[:50])

# Task 2 - Step 2 : Add Title and Labels
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")

# Task 2 - Step 3 : Display
plt.show()

```

```python
# TASK 3 : BAR PLOT (Vertical & Horizontal)
# Task 3 - Step 1 : Vertical Bar Chart
plt.bar(category[:10], sales[:10])

# Task 3 - Step 2 : Add Title & Labels
plt.title("Sales by Category - Vertical")
plt.xlabel("Category")
plt.ylabel("Sales")

# Task 3 - Step 3 : Show
plt.show()

# Task 3 - Step 1 : Horizontal Bar Chart
plt.barh(category[:10], sales[:10])

# Task 3 - Step 2 : Add Title & Labels
plt.title("Sales by Category - Horizontal")
plt.xlabel("Sales")
plt.ylabel("Category")

# Task 3 - Step 3 : Show
plt.show()
```

```python
# TASK 4 : MULTIPLE BAR PLOT

# Task 4 - Step 1 : Prepare Data
years = ["2021","2022","2023"]
online = [20000,25000,30000]
offline = [18000,22000,27000]

pos = [0,1,2]

# Task 4 - Step 2 : Create Multiple Bars
plt.bar([p-0.2 for p in pos], online, width=0.4, label="Online")
plt.bar([p+0.2 for p in pos], offline, width=0.4, label="Offline")

# Task 4 - Step 3 : Title & Labels
plt.title("Year Wise Sales Comparison")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.legend()

# Task 4 - Step 4 : Show
plt.show()
```

```python
# TASK 5 : STACKED BAR CHART
# Task 5 - Step 1 : Create Stacked Bar
plt.bar(years, online, label="Online")
plt.bar(years, offline, bottom=online, label="Offline")

# Task 5 - Step 2 : Add Labels
plt.title("Stacked Sales")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.legend()

# Task 5 - Step 3 : Show
plt.show()
```

```python
# TASK 6 : HISTOGRAM
# Task 6 - Step 1 : Create Histogram
plt.hist(quantity, bins=6)

# Task 6 - Step 2 : Labels
plt.title("Quantity Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

# Task 6 - Step 3 : Show
plt.show()
```

```python
# TASK 7 : PIE CHART
# Task 7 - Step 1 : Prepare Data
pie_values = sales[:4]
pie_labels = sub_category[:4]

# Task 7 - Step 2 : Create Pie Chart with %
plt.pie(pie_values, labels=pie_labels, autopct="%1.1f%%")

# Task 7 - Step 3 : Title
plt.title("Market Share by Sub Category")

# Task 7 - Step 4 : Show
plt.show()

```
